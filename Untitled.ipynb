```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import MinMaxScaler, RobustScaler, StandardScaler
from sklearn.model_selection import train_test_split
```

```python
dfWhole = pd.read_csv('delayed_flights_clean.csv')
#Transform DelayedMinutes to a binary variable
dfWhole['Delayed'] = dfWhole['DelayedMinutes'] > 0
dfWhole.drop(['Unnamed: 0', 'Date', 'DelayedMinutes'], axis = 1, inplace = True)
dfWhole.head()
```

At this point we could certainly transform the causes for delay to binary variables too, but it is interesting to derive a categorical feature from numerical variables (when resources allow it).

```python
def divideScaleConcat(df, dummyCol, targetCol, scaler):
    numeric = df.select_dtypes(include = np.number)
    scaled = scaleToDf(numeric, scaler)
    dummies = pd.get_dummies(df[dummyCol])
    target = df[targetCol]
    return pd.concat([scaled, dummies, target], axis = 1)

def scaleToDf(dfToScale, scaler):
    scaledDf = pd.DataFrame(scaler.transform(dfToScale), columns = dfToScale.columns)
    return scaledDf
```

```python
#sample the dataset to 15% of its original size
train, test = train_test_split(dfWhole, train_size = 0.15, 
                               stratify = dfWhole['UniqueCarrier'], shuffle = True,
                               random_state = 10) 
dfTrainAux = pd.DataFrame(train, 
                  columns = dfWhole.columns).reset_index(drop = True)
dfTestAux = pd.DataFrame(test,
                  columns = dfWhole.columns).reset_index(drop = True)
```

```python
scaler = RobustScaler()
#Get only the numeric columns from the train set
numericTrainDf = dfTrainAux.select_dtypes(include = np.number)
scaler.fit(numericTrainDf)
```

```python
trainDf = divideScaleConcat(dfTrainAux, 'UniqueCarrier', 'Delayed', scaler)
trainDf.info()
trainDf.head()
```

```python
trainDf.corr().iloc[:-1,-1].max()
```

```python
testDf = divideScaleConcat(dfTestAux, 'UniqueCarrier', 'Delayed', scaler)
testDf.info()
testDf.head()
```

#### Class exploration
So we have 2 classes basically: Delayed True and Delayed False. Let's explore if there is balance between these.

```python
i = trainDf['Delayed'].sum()*100 / trainDf.shape[0]
print(f'{i:.2f}% of delayed flights on the train sample')
```

We have no grave unbalance between classes, so we will not need to use Under or over sampling in this particular case.

#### kFolds 
We use a 5 fold cross-validation scheme

```python
from sklearn.model_selection import KFold

folds = KFold(n_splits = 5, shuffle = True, random_state = 3)
```

#### RandomForest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
xTrain = trainDf.iloc[:,:-1]
yTrain = trainDf.iloc[:,-1]

hyperParams = [{'n_estimators' : list(range(10,110,10)),
               'max_depth' : list(range(3,6))
               }]

clf = RandomForestClassifier(min_samples_split = 0.1,
                            n_jobs = -1, random_state = 1)

model = GridSearchCV(estimator = clf, param_grid = hyperParams, cv = folds,
                    scoring = ['roc_auc', 'f1', 'precision', 'recall', 'accuracy'],
                    refit = 'roc_auc')
                    
model.fit(xTrain, yTrain)
```

```python
cvResults = pd.DataFrame(model.cv_results_)
cvResults
```

```python
bestForest = model.best_estimator_
print(bestForest)
```

The best overall estimator would be the one ranked with 1. We get the one ranked with 6 because of the refit auc_roc parameter. Let's visualize a tree of this forest.

```python
from sklearn.tree import export_graphviz
from six import StringIO
from IPython.display import Image  
import pydotplus

dot_data = StringIO()
export_graphviz(bestForest[0], out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True,feature_names = trainDf.columns[:-1],
                class_names=['Not Delayed','Delayed'])
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
graph.write_png('tree.png')
Image(graph.create_png())
```

Let's build a second forest, with the rank 1 forest which has depth 5 and 40 estimators, and compare their roc curves.

```python
hyperParams = [{'n_estimators' : [40],
               'max_depth' : [5]
               }]

clf = RandomForestClassifier(min_samples_split = 0.1,
                            n_jobs = -1, random_state = 1)

aspirantModel = GridSearchCV(estimator = clf, param_grid = hyperParams, 
                             cv = folds, scoring = 'roc_auc')

aspirantModel.fit(xTrain, yTrain)

aspirantForest = aspirantModel.best_estimator_
```

```python
from sklearn import metrics

yTest = testDf.iloc[:,-1]
xTest = testDf.iloc[:,:-1]

bestYPred = bestForest.predict(xTest)
print('Confusion matrix for best forest', metrics.confusion_matrix(yTest, bestYPred))

aspirantYPred = aspirantForest.predict(xTest)
print('Confusion matrix for aspirant forest', metrics.confusion_matrix(yTest, aspirantYPred))
```

The results are fairly similar, but the aspirantForest has a much lower rate of False Negatives (flights that actually arrived late but were predicted on time) than the 'best forest'. True negatives stay more or less the same, while true positives (predicted as delayed and actually delayed) go up for the aspirant forest. All in all, it seems our aspirantForest is the winner of this round.

```python
def plotROC(classifier, xTest, yTest, ax, label):
    
    yPredProba = classifier.predict_proba(xTest)[:,1]
    fpr, tpr, _ = metrics.roc_curve(yTest, yPredProba)
    auc = metrics.roc_auc_score(yTest, yPredProba)
    
    ax.plot(fpr, tpr, label = label + f', auc = {auc:.6f}')
```

```python
fig, ax = plt.subplots(figsize = (12,8))
plotROC(bestForest, xTest, yTest, ax, 'Best Forest')
plotROC(aspirantForest, xTest, yTest, ax, 'Aspirant Forest')
ax.legend()
ax.grid()
```

The performance of our aspirantForest is slightly better, but not as much as anticipated by just looking at the confusion matrix. Finally we can look at the accuracy of the model

```python
print('Our model has an accuracy score of:', metrics.accuracy_score(yTest, aspirantYPred))
```

A dummy model would have an accuracy of approximately 60% (% of population of majoritary class). All in all, a good model.

#### XGBoost

```python
from sklearn import ensemble

fixedParams = {'max_iter' : 1200, 'early_stopping' : True, 'validation_fraction' : 0.1, 
              'n_iter_no_change' : 5, 'tol' : 0.01, 'random_state' : 0,
              'loss' : 'binary_crossentropy'}

paramDist = {'max_depth' : list(range(2,11)), 'learning_rate' : np.array(range(1, 11, 1))/10}

gradBoost = ensemble.HistGradientBoostingClassifier(**fixedParams)

CVclf = GridSearchCV(estimator = gradBoost, param_grid = paramDist, cv = folds,
                 scoring = ['roc_auc', 'f1', 'precision', 'recall', 'accuracy'],
                 refit = 'roc_auc', verbose = 1)

CVclf.fit(xTrain, yTrain)
```

```python
pd.DataFrame(CVclf.cv_results_)
```

A downside of having so much estimators, is that it becomes increasingly difficult to manually select one. However, the difference between our best manual selection and best_estimator_ should be minimal.

```python
clfXGB = CVclf.best_estimator_
yPred = clfXGB.predict(xTest)

print('Confusion matrix for XGBoost', metrics.confusion_matrix(yTest, yPred))
print('Confusion matrix for Aspirant Forest', metrics.confusion_matrix(yTest, aspirantYPred))
```

Even though XGBoost will have more on time flights predicted as delayed, it will have much less actual delayed flights that are predicted on time. By looking at the confusion matrix, it seems one model will perform better than the other depending on what we want to minimize. To assess overall performance, we can look at the ROC curve:

```python
fig, ax = plt.subplots(figsize = (12,8))
plotROC(clfXGB, xTest, yTest, ax, 'XGBoost')
plotROC(aspirantForest, xTest, yTest, ax, 'Aspirant Forest')
ax.set_title('ROC curve XGBoost vs RandomForest')
ax.legend()
ax.grid()
```

It seems as the XGBoost model will beat our aspirant random forest, but margins are really low.

#### Logistic Regression
For this particular model, I will try something new. In my case, the dataset has been tailored so that the direct causes of delay have a low correlation with the actual delay. However, in previous tasks, I proved that the sum of all of these direct causes has a high correlation with the delay. Let us make a Logistic regression with only one predictor consisting of the sum of these direct causes for delay. This procedure should be similar to a PCA, but less black boxy-ish. The drawback is that we will lose information in the process.

```python
#Our variables dfTestAux and dfTrainAux have been left unscaled 
#and untouched purposely for this last bit.
delayCauses = ['NASDelay', 'WeatherDelay', 'SecurityDelay', 
               'CarrierDelay', 'LateAircraftDelay']
dfTrainAux['DelaySum'] = dfTrainAux[delayCauses].sum(axis = 1)
dfTestAux['DelaySum'] = dfTestAux[delayCauses].sum(axis = 1)

dfTrainAux.head()
```

```python
def scaleConcat(df, predictor, target, scaler):
    
    arrayToScale = np.array(df[predictor]).reshape(-1,1)
    predictorSeries = scaleToSeries(arrayToScale, predictor, scaler)
    return pd.concat([predictorSeries, df[target]], axis = 1)    

def scaleToSeries(s, name, scaler):
    ss = scaler.transform(s)
    return pd.Series(ss.flatten(), name = name)
```

```python
scaler = RobustScaler()
target = 'Delayed'
predictor = 'DelaySum'
#Only scale the predictor and target columns, no need for dummies
scaler = RobustScaler()
scaler.fit(np.array(dfTrainAux[predictor]).reshape(-1,1))

trainDf = scaleConcat(dfTrainAux, predictor, target, scaler)
testDf = scaleConcat(dfTestAux, predictor, target, scaler)

trainDf.head()
```

```python
from sklearn.linear_model import LogisticRegression

paramDist = {'solver' : ('newton-cg', 'lbfgs', 'liblinear')}
#Somehow with 'saga' and 'sag' solvers, the model is not converging

logReg = LogisticRegression()

LRclf = GridSearchCV(estimator = logReg, param_grid = paramDist, cv = folds,
                 scoring = ['roc_auc', 'f1', 'precision', 'recall', 'accuracy'],
                 refit = 'roc_auc', verbose = 1)

LRclf.fit(np.array(trainDf[predictor]).reshape(-1,1),
          trainDf[target]
         )
```

```python
bestLR = LRclf.best_estimator_
yPredLR = bestLR.predict(np.array(testDf[predictor]).reshape(-1,1))

print('Confusion matrix for Logistic Regression', 
      metrics.confusion_matrix(testDf[target], yPredLR)
     )
```

```python
fig, ax = plt.subplots(figsize = (12,8))
plotROC(clfXGB, xTest, yTest, ax, 'XGBoost')
plotROC(bestLR, np.array(testDf[predictor]).reshape(-1, 1),
        testDf[target], ax, 'Logistic Regression'
       )
ax.set_title('ROC curve XGBoost vs Logistic Regression')
ax.legend()
ax.grid()
```

This curve may seem overfitted, but we should bear in mind that we are crossvalidating and tuning hyperparameters. It is in fact a very naive reasoning: If the sum of the causes for the delay are greater than 0, then the flight is delayed. On the other hand, if there are no causes for the delay, there is no delay.  
The huge success is due to a convenient dimensionality reduction. However, this transformation comes at the cost of not including the specific cause for the delay, so even though we can perfectly predict whether a flight will be late or not, the model alone cannot help us improve flight punctuality. This is an easy fix though, like for example plotting the causes for delay in a separate figure.
